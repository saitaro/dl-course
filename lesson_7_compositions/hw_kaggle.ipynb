```python
# Будем работать с каждым значением PaymentMethod по отдельности
origin_cat_cols = cat_cols.copy()

cat_cols.remove("PaymentMethod")
cat_cols.extend(pd.get_dummies(df.PaymentMethod, drop_first=True).columns)
```

```python
def prepare_data(df: pd.DataFrame) -> pd.DataFrame:
    df["TotalSpent"] = pd.to_numeric(df.TotalSpent, errors="coerce")
    df["TotalSpent"] = df["TotalSpent"].fillna(df.TotalSpent.mean())

    df["Sex"] = df.Sex.eq("Male")
    df["HasPartner"] = df.HasPartner.eq("Yes")
    df["HasChild"] = df.HasChild.eq("Yes")
    df["HasPhoneService"] = df.HasPhoneService.eq("Yes")
    df["HasMultiplePhoneNumbers"] = df.HasMultiplePhoneNumbers.eq("Yes")
    df["HasInternetService"] = df.HasInternetService.map(
        {
            "No": 0,
            "DSL": 1,
            "Fiber optic": 2,
        }
    )
    df["HasOnlineSecurityService"] = df.HasOnlineSecurityService.eq("Yes")
    df["HasOnlineBackup"] = df.HasOnlineBackup.eq("Yes")
    df["HasDeviceProtection"] = df.HasDeviceProtection.eq("Yes")
    df["HasTechSupportAccess"] = df.HasTechSupportAccess.eq("Yes")
    df["HasOnlineTV"] = df.HasOnlineTV.eq("Yes")
    df["HasMovieSubscription"] = df.HasMovieSubscription.eq("Yes")
    df["HasContractPhone"] = df.HasContractPhone.map(
        {
            "Month-to-month": 0,
            "One year": 1,
            "Two year": 2,
        }
    )
    df["IsBillingPaperless"] = df.IsBillingPaperless.eq("Yes")

    payment_method_dummies = pd.get_dummies(df.PaymentMethod, drop_first=True)
    df = df.drop(columns="PaymentMethod")
    df = pd.concat([df, payment_method_dummies], axis=1)

    return df
```

```python
df = prepare_data(df)
df_control = prepare_data(df_control)
```
